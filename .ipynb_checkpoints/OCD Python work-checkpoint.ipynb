### Alternative method

```python
def hue(x1, hue1, title, xlabel, ylabel):

#     obsession_type = ocd_data['Obsession Type']
#     gender = ocd_data['Gender']

    plt.figure(figsize = (10,7))

    sns.countplot(x=x1, hue=hue1)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
obsession_type = ocd_data['Obsession Type']
gender = ocd_data['Gender']

hue(obsession_type, gender, "Number of Patients based on Gender from the resulting Compulsion Type", "Compulsion Type", "Number of Patients")
```
